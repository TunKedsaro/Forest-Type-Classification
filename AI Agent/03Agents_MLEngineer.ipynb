```python
def fn(ip):
    prompt = ''' You are a Data Engineer with over 20 years of experience. Today, I need your help with creating a Python function based on the index feature I provide, to generate new features for this dataset. Here is an example dataset:
    data = {
        'id': [2002, 3212, 13312],
        '01Coastal': [293, 197, 929],
        '12SWIR1': [1927, 1598, 1975],
        '13SWIR2': [1038, 697, 1031],
        '02Blue': [278, 201, 982],
        '03Green': [475, 347, 1020],
        '04Red': [453, 228, 856],
        '05Veg5': [987, 682, 1220],
        '06Veg6': [1773, 1982, 2051],
        '07Veg7': [2184, 2449, 2421],
        '08NIR': [1900, 2254, 2392],
        '09NIRn': [2343, 2685, 2671],
        '10WV': [3039, 2690, 2683],
        'nforest_type': ['MDF', 'DDF', 'MDF']
    }

    instruction :
    - Create a Python function that can be used immediately with the DataFrame df
    - Provide only the function, as shown in the example.
    - The output should not have these (backticks, explanation, opinion, quotes)

    Example:
    # Input:
    NDVI (Normalized difference vegetation index) : (NIR - RED) / (NIR + RED)
    # Output:
    df['NDVI'] = (df['08NIR'] - df['04Red']) / (df['08NIR'] + df['04Red'])

    # Input:
    <input></input>
    # Output '''
    prompt = prompt.replace("<input></input>",ip)
    return prompt
```
